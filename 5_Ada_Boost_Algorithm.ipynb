# Ada -Boost Algorithm


- **Adaptive boosting is a boosting ensemble learning that sequentially trains weak classifier to create a strong classifier**

- **Properties:**
  1. Ada Boost trains weak learners sequentially
  2. Emphasizes on getting misclassified samples right
  3. Weighted Gini Index is used to decide which feature will be used
  4. Ada Boost trains weak learners sequentially
  5. At each iteration,we boost weights of misclassified samples
  6. Not best suited to work with Noisy data

#data preprocessing


```python
#Missing value
df.isnull().sum()
```

```python
df = df.drop('Cabin',axis=1)
```

```python
df.columns
```

```python
from sklearn.impute import SimpleImputer
print(f"Number of Null values in Age Column before imputation: {df['Age'].isnull().sum()}")

# for int values
si = SimpleImputer(missing_values=np.nan,strategy='mean')
df['Age'] = si.fit_transform(df[['Age']])
print(f"Number of Null values in Age Column after imputation: {df['Age'].isnull().sum()}")
```

```python
df['Survived'] = si.fit_transform(df[['Survived']])

df['Fare'] = si.fit_transform(df[['Fare']])
```

```python
#for object(str)
si1 = SimpleImputer(missing_values=np.nan,strategy='most_frequent')
df[['Embarked']] = si1.fit_transform(df[['Embarked']])
```

```python
df.isnull().sum()
```

```python
df = df.drop(['Name','Ticket'],axis = 1)
```

```python
df
```

```python
#outlier detection
thershold = 3
from scipy.stats import zscore

numerical_feature =['Fare','Age']
z_score = zscore(df[numerical_feature])
print(z_score)
```

```python
outliers = np.where(z_score > thershold)
print(outliers)
```

```python
from scipy.stats.mstats import winsorize
df['Age'] = winsorize(df['Age'],limits=[0.15,0.15])
df['Fare'] = winsorize(df['Fare'],limits=[0.15,0.15])
```

```python
#Encoding
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder()
df['Sex'] = oe.fit_transform(df[['Sex']])
df['Embarked'] = oe.fit_transform(df[['Embarked']])

df['Survived'] = oe.fit_transform(df[['Survived']])
```

```python
df
```

```python
X = df.drop('Survived', axis = 1)
y = df['Survived']
```

# train test split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

# Ada-Boost

```python
from sklearn.ensemble import AdaBoostClassifier
ab = AdaBoostClassifier(n_estimators=45,learning_rate=0.01,random_state=0)
ab.fit(X_train,y_train)
y_pred = ab.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print('Accuracy :',accuracy_score(y_test,y_pred))
```
